```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
base_path = "/content/drive/MyDrive/ImageProcessingHomework/"
file_names = ["coins.ascii.pgm", "fruit.pgm", "saturn.ascii.pgm"]
```

```python
def adjust_contrast(image_array, contrast_value):
    adjusted_array = np.clip(image_array + contrast_value, 0, 255)
    return adjusted_array.astype(np.uint8)
```

```python
for file_name in file_names:
    if file_name == ".DS_Store":
        continue  # Skip .DS_Store file

    image_path = base_path + file_name
    contrast_value = 50  # Replace with the desired contrast adjustment value

    # Load the image
    input_image = Image.open(image_path).convert('L')
    image_array = np.array(input_image)

    # Adjust contrast
    adjusted_array = adjust_contrast(image_array, contrast_value)

    # Display the original and adjusted images
    print(f"Original Image: {file_name}")
    plt.imshow(image_array, cmap='gray')
    plt.axis('off')
    plt.show()

    print(f"Adjusted Image: {file_name}")
    plt.imshow(adjusted_array, cmap='gray')
    plt.axis('off')
    plt.show()
```
